```python
total_class = float(input('Number of the Total Class: '))
total_attendance = float(input('Number of total Class Attendance: '))
attendance_percentage = (total_attendance/total_class)*100
```

Number 04

```python
marks = float(input('Enter the Marks: '))
```

```python
if marks<25:
    print('Your Grade is F')
elif marks>=25 and marks<=44:
    print('Your Grade is E')
elif marks>=45 and marks<=49:
    print('Your Grade is D')
elif marks>=50 and marks<=59:
    print('Your Grade is C')
elif marks>=60 and marks<=79:
    print('Your Grade is B')
elif marks>=80 and marks<=89:
    print('Your Grade is A')
elif marks>=90:
    print('Your Grade is A+')
else:
    print('Error')
```
